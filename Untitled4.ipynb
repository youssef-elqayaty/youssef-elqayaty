let's creating our dash board

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

# ====== 1. Load Data ======
df = pd.read_csv("flights_clean.csv")

# Convert date
df["FlightDate"] = pd.to_datetime(df["FlightDate"])

# Create additional columns
df["DayName"] = df["FlightDate"].dt.day_name()
df["MonthName"] = df["FlightDate"].dt.month_name()
df["Hour"] = pd.to_datetime(df["CRSDepTime"], format="%H:%M:%S", errors="coerce").dt.hour

# ====== 2. Create App ======
app = Dash(__name__)
app.title = "Flight Insights Dashboard"

# ====== 3. Layout ======
app.layout = html.Div([
    html.H1("✈️ Flight Analytics Dashboard", style={"textAlign": "center"}),

    # --- Filters ---
    html.Div([
        html.Label("Select Airline:"),
        dcc.Dropdown(
            id="carrier_filter",
            options=[{"label": i, "value": i} for i in sorted(df["UniqueCarrier"].unique())],
            value=[],
            multi=True,
            placeholder="Select one or more airlines"
        ),
        html.Br(),

        html.Label("Select Month:"),
        dcc.Dropdown(
            id="month_filter",
            options=[{"label": m, "value": m} for m in sorted(df["MonthName"].unique())],
            value=[],
            multi=True,
            placeholder="Select one or more months"
        ),
        html.Br(),

        html.Label("Select Destination:"),
        dcc.Dropdown(
            id="dest_filter",
            options=[{"label": d, "value": d} for d in sorted(df["Dest"].unique())],
            value=[],
            multi=True,
            placeholder="Select one or more destinations"
        )
    ], style={"width": "60%", "margin": "auto"}),

    html.Br(),

    # --- Graphs ---
    dcc.Tabs([
        dcc.Tab(label="📈 Line Chart (Average delay over time)", children=[dcc.Graph(id="line_chart")]),
        dcc.Tab(label="🥧 Pie Chart (Flight share by airline)", children=[dcc.Graph(id="pie_chart")]),
        dcc.Tab(label="🔥 Heatmap (Best booking times)", children=[dcc.Graph(id="heatmap_chart")]),
        dcc.Tab(label="📊 Bar Chart (Average delay by airline)", children=[dcc.Graph(id="bar_chart")])
    ]),

    html.Br(),

    # --- Insights Section ---
    html.Div(id="insights_box", style={
        "border": "2px solid #007BFF",
        "borderRadius": "10px",
        "padding": "20px",
        "margin": "20px",
        "backgroundColor": "#F8F9FA"
    })
])

# ====== 4. Callbacks ======
@app.callback(
    [Output("line_chart", "figure"),
     Output("pie_chart", "figure"),
     Output("heatmap_chart", "figure"),
     Output("bar_chart", "figure"),
     Output("insights_box", "children")],
    [Input("carrier_filter", "value"),
     Input("month_filter", "value"),
     Input("dest_filter", "value")]
)
def update_dashboard(carriers, months, dests):
    dff = df.copy()

    # Airline filter
    if carriers:
        dff = dff[dff["UniqueCarrier"].isin(carriers)]

    # Month filter
    if months:
        dff = dff[dff["MonthName"].isin(months)]

    # Destination filter
    if dests:
        dff = dff[dff["Dest"].isin(dests)]

    # If data is empty after filtering
    if dff.empty:
        return {}, {}, {}, {}, html.Div([
            html.H3("⚠️ No data available for the selected filters")
        ])

    # --- Line Plot ---
    line_data = dff.groupby("FlightDate")["DepDelay"].mean().reset_index()
    line_fig = px.line(line_data, x="FlightDate", y="DepDelay",
                       title="Average Departure Delay Over Time")

    # --- Pie Chart ---
    pie_fig = px.pie(dff, names="UniqueCarrier", title="Airline Share of Total Flights")

    # --- Heatmap ---
    heatmap_data = dff.groupby(["DayName", "Hour"])["DepDelay"].mean().reset_index()
    heatmap_fig = px.density_heatmap(heatmap_data, x="Hour", y="DayName", z="DepDelay",
                                     color_continuous_scale="Blues",
                                     title="Best Times (Lowest Average Delay)")

    # --- Bar Chart ---
    bar_data = dff.groupby("UniqueCarrier")["DepDelay"].mean().reset_index()
    bar_fig = px.bar(bar_data, x="UniqueCarrier", y="DepDelay",
                     title="Average Departure Delay by Airline")

    # --- Insights ---
    if not heatmap_data.empty:
        best_time = heatmap_data.loc[heatmap_data["DepDelay"].idxmin()]
        insights = html.Div([
            html.H3("🔍 Insights & Recommendations"),
            html.P(f"✅ Best time to travel is {best_time['DayName']} at {int(best_time['Hour'])}:00"),
            html.P("✅ Early morning flights are often less delayed."),
            html.P("✅ Some airlines have lower average delays → better to choose them.")
        ])
    else:
        insights = html.Div([html.H3("⚠️ Not enough data to provide insights.")])

    return line_fig, pie_fig, heatmap_fig, bar_fig, insights

# ====== 5. Run App ======
if __name__ == "__main__":
    app.run(debug=True)

```
